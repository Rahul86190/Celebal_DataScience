```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Titanic_cleaned_data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
plt.plot(df["age"])
```

```python
plt.bar(df["class"].value_counts().keys(),df["class"].value_counts().values)
```

```python
plt.bar(df["sex"].value_counts().keys(),df["sex"].value_counts().values)
```

```python
df["survived"].value_counts().keys().astype(int).to_list()
```

```python
plt.bar(df["survived"].value_counts().keys().astype(int).to_list(),df["survived"].value_counts().values)
plt.xticks(df["survived"].value_counts().keys().astype(int).to_list())
plt.show()
```

```python
plt.bar(df["deck"].value_counts().keys(),df["deck"].value_counts().values,color="red")
```

```python
plt.pie(df["sex"].value_counts().values,labels=df["sex"].value_counts().keys(),autopct="%2.2f%%",explode=[0.1,0],shadow="B")
plt.plot()
```

```python
df.head()
```

```python
plt.pie(df["alive"].value_counts().values,labels=df["alive"].value_counts().keys(),autopct="%2.2f%%",wedgeprops={"ec":"black","ls":"dashed"},explode=[0,.04])
plt.title("Alive",c="Red")
plt.show()
```

```python
plt.hist(df["age"],edgecolor="black",color="green")
plt.show()
```

```python
plt.boxplot(df["age"])
plt.show()
```

```python
sns.catplot(x=df["sex"],y=df["age"],kind='bar',data=df)   
plt.xticks(rotation=20)
plt.show()
```

```python
sns.countplot(data=df, x='survived',palette='Set2')
plt.xlabel("Survival")
plt.ylabel("Number of people")
plt.show()
```

```python
sns.countplot(data=df, x='sex', hue='survived', palette='coolwarm')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.countplot(data=df, x='class', hue='survived', palette='viridis')
```

```python
plt.hist(df["age"].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
sns.boxplot(data=df, x='survived', y='age')
plt.grid(axis='y', linestyle='--')
```

```python
sns.countplot(data=df, x='embarked', hue='survived')
```

```python
sns.violinplot(data=df, x='class', y='age', hue='sex')
plt.show()
```

```python
sns.pairplot(df, hue="survived", palette="Set1")  # best plot 
plt.show()
```
